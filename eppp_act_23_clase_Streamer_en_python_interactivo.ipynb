# Streamer
![](https://raw.githubusercontent.com/zavalaalexia128/EPPP-Act-23-clase-nombre-en-python-interactivo/refs/heads/main/infoclasestreamer.png)

```python
class Streamer:
    def __init__(self, nickname, plataforma, categoria):
        self.nickname = nickname
        self.plataforma = plataforma
        self.categoria = categoria
        self.en_vivo = False
        self.espectadores = 0

    def iniciar_stream(self):
        self.en_vivo = True
        self.espectadores = 100
        print(f"\n[SISTEMA]: ¡{self.nickname} está ahora EN VIVO en {self.plataforma}!")
        print(f"Transmitiendo: {self.categoria}")

    def mostrar_info(self):
        estado = "EN VIVO" if self.en_vivo else "OFFLINE"
        print(f"\n--- PERFIL DE STREAMER ---")
        print(f"Nickname: {self.nickname}")
        print(f"Plataforma: {self.plataforma}")
        print(f"Categoría: {self.categoria}")
        print(f"Estado actual: {estado}")
        if self.en_vivo:
            print(f"Espectadores: {self.espectadores}")
        print("-" * 26)

print("¡Bienvenido al Simulador de Streamers!")
print("Por favor, configura tu canal:")

nombre = input("¿Cuál es tu Nickname?: ")
sitio = input("¿En qué plataforma vas a transmitir? (Twitch/YouTube/Kick): ")
tema = input("¿Qué contenido vas a streamear? (Gaming/Just Chatting/Cooking): ")

mi_streamer = Streamer(nombre, sitio, tema)

mi_streamer.mostrar_info()

opcion = input(f"\n¿Deseas iniciar el directo ahora, {nombre}? (si/no): ").lower()

if opcion == "si":
    mi_streamer.iniciar_stream()
    mi_streamer.mostrar_info()
else:
    print(f"\nEntendido. {nombre} se queda en modo descanso por hoy.")

print("\n" + "="*30)
print("Programa terminado")
print("Programa realizado por Zavala Medina NL= 54")
```
